```python
Question 9 
```

```python
def fact(n):
    f = 1
    for i in range(1, n+1):
        f *= i
    return f

n = int(input("Enter number: "))
s = 0
temp = n

while temp > 0:
    d = temp % 10
    s += fact(d)
    temp //= 10

print("Strong Number" if s == n else "Not Strong")
```

```python
n = int(input("Enter number: "))
i = 2
factors = []

while i * i <= n:
    if n % i == 0:
        factors.append(i)
        while n % i == 0:
            n //= i
    i += 1

if n > 1:
    factors.append(n)

print("Prime Factors =", factors)
```

```python

def pascal_pyramid(n):
    for i in range(n + 1):
        print(" " * (n - i), end="")
        val = 1
        for j in range(i + 1):
            print(val, end=" ")
            val = val * (i - j) // (j + 1)
        print()

n = int(input())
pascal_pyramid(n)
```

```python
Question 22
```

```python
def string_pattern(s):
    n = len(s)
    for i in range(n):
        left = s[:n - i]
        right = left[::-1]
        print(left + " " * i + right)
    for i in range(n - 2, -1, -1):
        left = s[:n - i]
        right = left[::-1]
        print(left + " " * i + right)

s = input()
string_pattern(s)
```
